# Irradiated vs Nonirradation Chow

## Experimental Design

Three cages (8 mice total) of normally housed conventional mice on the standard irradiated 5L0D chow from picolab were monitored for 5 days, after which they were switched to a nonirradiated 5001 rodant chow which was identical to 5L0D in ingrediants.  Mice were kept on this diet for 10 days afterwhich they were returned to their normal diet.

## Analysis

The analysis begins after the the cluster.split was run on mothur. At this point it was necessary to correct some of the typos which were present in the sample names.  This was done using the python script below.

```python
%install_ext https://raw.github.com/kdiverson/ipython-mothurmagic/master/mothurmagic.py
```

```python
%load_ext mothurmagic
```

```python
file=open('chow.final.counts','r')
filedata = file.read()
file.close()

newdata = filedata.replace('261_14\t', '26_1_14\t')
newdata = newdata.replace('26_1\t', '26_1_16\t')
newdata = newdata.replace('2711_16\t', '271_1_16\t')
newdata=newdata.replace('271_3_17\t', '271_3_14\t')
newdata=newdata.replace('Chow\t','Chow_NA_NA\t')

file = open('chow.final.counts','w')
file.write(newdata)
file.close()
```

the mothur pipeline was then used to analyze beta diversity in the sample by measuring theytayc distances and plotting this data in an ordination.  Some of Niel's samples begining in '2779' were removed prior to this analysis and samples were subsampled at 1111 to accomodate the nonirradiated chow which was sequenced.

```python
%%mothur
make.shared(list=./chow.final.an.unique_list.list, count=./chow.final.counts, label=0.03)
classify.otu(list=./chow.final.an.unique_list.list, count=./chow.final.counts, taxonomy=./chow.final.taxonomy, label=0.03)
count.groups(shared=./chow.final.an.unique_list.shared)
remove.groups(shared=./chow.final.an.unique_list.shared, groups=2779_STK-2779_LIQ-2779)
sub.sample(shared=./chow.final.an.unique_list.0.03.pick.shared, size=1111)
dist.shared(shared=./chow.final.an.unique_list.0.03.pick.0.03.subsample.shared, calc=thetayc)
nmds(phylip=chow.final.an.unique_list.0.03.pick.0.03.subsample.thetayc.0.03.lt.dist)
```

The axes file was then manipulated into a metadata file including the cage, mouse and day data by replacing the underscores in the names with tabs.

```python
file = open('chow.final.an.unique_list.0.03.pick.0.03.subsample.thetayc.0.03.lt.nmds.axes','r')
meta = open('chow.metadata','w')

meta.write('Name\tCage\tMouse\tDay\tAxis1\tAxis2\n')

next(file)

for line in file:
    name = line.split('\t')[0]    
    line = line.replace('_','\t')
    line = name + '\t' + line
    meta.write(line)
    print(line)
  
meta.close
file.close

```

This metadata file was used to create an nmds plot in R as below

```python
%load_ext rmagic
```

```python
%%R
x<-read.table('~/Desktop/chow/chow.metadata', header=T)
nonirradiated <- x[x$Day %in% c(6:15), c(5,6)]
irradiated <- x[x$Day %in% c(1:5,16:20), c(5,6)]
irrChow <- x[x$Cage=='IrradiatedChow', c(5,6)]
nonirrChow <- x[x$Cage=='NonirradiatedChow', c(5,6)]


plot(x$Axis1,x$Axis2, xlim=c(-1.4,1.4), ylim=c(-1,1), xlab='NMDS Axis 1',ylab='NMDS Axis 2')


points(nonirradiated, pch=21, col='black', bg='red', cex=1.5)
points(irradiated, pch=21, col='black', bg='chartreuse4', cex=1.5)
points(irrChow, pch=24, col='black', bg='chartreuse4', cex=1.5)
points(nonirrChow, pch=24, col='black', bg='red', cex=1.5)
abline(v=0.3)
abline(0.4,0)

legend('topright',legend=c('nonirrad  ','irrad  ','nonirrchow  ','irrchow  '), col='black', 
       pch=c(21,21,24,24), pt.cex=1.5, pt.bg=c('red','chartreuse4','red','chartreuse4'))
```

Look at those nice colors.  Initially it doesn't seem like there is a difference between the two groups they cluster rather tightly.  But what are the points that seem dispersed about the top of the plot?  I've added line to help isolate them and given these cut offs it sems the samples are these here:

```python
%%R
losers = x[x$Axis1>0.3 & x$Axis2>0.4, c(1)]
print(losers)
```

```python
%%mothur 
get.relabund(shared=chow.final.an.unique_list.0.03.pick.0.03.subsample.shared)
```

```python
%%R
relabund<-read.table('~/Desktop/Chow/chow.final.an.unique_list.0.03.pick.0.03.subsample.relabund',header=T)
cluster <- x$Name[!(x$Name %in% losers)]
clusterOTU <- relabund[relabund$Group %in% c(cluster,'NonirradiatedChow_NA_NA','IrradiatedChow_NA_NA'), c(4:length(relabund))]
loserOTU <- relabund[relabund$Group %in% losers, c(4:length(relabund))]
clustermeans <- colMeans(clusterOTU)
losermeans <- colMeans(loserOTU)
clustermeans <-clustermeans[clustermeans>0.01]
losermeans <- losermeans[losermeans>0.01]
print(clustermeans)
print(losermeans)

```
